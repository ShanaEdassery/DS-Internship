```python
import pandas as pd
import numpy as np

data2 = {
    'Employee': ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve', 'Bob'],
    'Department': ['Sales', 'HR', None, 'Tech', 'HR', 'HR'],
    'Salary': [50000, 40000, 55000, None, 42000, 40000],
    'Performance': [88, 72, None, 95, 68, 72]
}

df2 = pd.DataFrame(data2)

# Fix missing values
df2['Salary']=df2['Salary'].fillna(df2['Salary'].mean())
df2['Performance']=df2['Performance'].fillna(df2['Performance'].mean())
df2['Department']=df2['Department'].fillna('Unknown')

# Remove duplicates
df2=df2.drop_duplicates()

print(df2)

print('shape=', df2.shape)
print('\nData types:')
print(df2.dtypes)
print('\n"Basic Stats"')
print(df2.describe())

import seaborn as sns
import matplotlib.pyplot as plt

sns.barplot(data=df2, x='Employee', y='Salary', color='lightblue')
plt.title('employees salaries')
plt.show()
```

```python
sns.barplot(data=df2, x='Employee', y='Performance', color='silver')
plt.title('employee performance')
plt.show()
```

```python
fig, axes= plt.subplots(1,2,figsize=(12,4))
sns.barplot(data=df2, x='Employee',y='Salary', ax=axes[0])
axes[0].set_title('Salaries')

sns.barplot(data=df2, x='Employee', y='Performance', ax=axes[1])
axes[1].set_title('Performance')

plt.tight_layout()
plt.show()
```
